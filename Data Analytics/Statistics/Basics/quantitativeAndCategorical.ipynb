First, we convert it to a pandas dataframe, and we would constrcut boxplot and bar plot to examine the quantitative side and qualitative side of the data. We decide the quantitative to contain just the data itself, and we use a floor function to get the integer part of the number and make that our categorical data. 

```python
# Set the range of the temperature data
rangeOfTemp = np.floor(bodyTemp)

# Convert to pandas dataframe
tempDF = pd.DataFrame()
tempDF["temp"] = bodyTemp
tempDF["tempRange"] = rangeOfTemp
tempDF["tempRange"] = tempDF["tempRange"].astype("str")
# Make a summary of body temperature
tempDF["temp"].describe()
# Make a summary of body temperature range
tempDF["tempRange"].describe()
```

```python
tempDF["tempRange"].value_counts().plot(kind="bar")
```

```python
# The dots in the graph gives us the potential outliers
# The bottom line is the min
# The second line is first 25%
# The third line is middle
# The forth line is 75%
# The top line is max
tempDF["temp"].plot(kind="box")
```

Let's take a look at what would happen if we group them all by their range and make a graph about their comparison

```python
def boxPlot(data, color):
    data["temp"].plot(kind="box")
grid = sns.FacetGrid(tempDF, col="tempRange")
grid.map_dataframe(boxPlot)
```
