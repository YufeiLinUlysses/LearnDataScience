# Bivariate Data
Data for variables are known as bivariate data.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
manatees = pd.read_csv("./DataSource/manatees.csv")
manatees.head(3)
```

YEAR: The year we collect manatees data from

Boats: Number of boats registered in Fl (in 1,000s of boats)

Manatees: Number of manatees killed, with cause determined to be boats.

```python
plt.scatter(x="Year",y="Boats",data=manatees)
plt.title("Year-Boats")
plt.xlabel("Year")
plt.ylabel("Boats")
```

```python
plt.scatter(x="Boats",y="Manatees",data=manatees)
```
