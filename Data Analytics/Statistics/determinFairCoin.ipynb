Fair Dice

```python
# Get numpy and all other important modules
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
%matplotlib inline
# Create a fair dice with six sides
coin = np.arange(2)
coin
# Fetch 1
np.random.choice(coin)
# Fetch 2, with duplicates
np.random.choice(coin, size=2, replace=True)
# Fetch 2, without duplicates
np.random.choice(coin, size=2, replace=False)
# check possibility
sum(np.random.choice(coin,size=500))/500
# double check
doubleCheck = np.array([sum(np.random.choice(coin,size=500))/500 for _ in range(1000)])
_ = plt.hist(doubleCheck, bins='auto')
plt.title("Histogram of Fair Coin probability")
plt.xlim([0,1])
plt.show()
```

```python
# Unfair Dice
unfair = np.array([0] + [1]*10)
unfair
# Fetch 1
np.random.choice(unfair)
# Fetch 2, with duplicates
np.random.choice(unfair, size=2, replace=True)
# Fetch 2, without duplicates
np.random.choice(unfair, size=2, replace=False)
# check possibility
sum(np.random.choice(unfair,size=500))/500
# double check
doubleCheck = np.array([sum(np.random.choice(unfair,size=500))/500 for _ in range(1000)])
_ = plt.hist(doubleCheck, bins='auto')
plt.title("Histogram of Unfair Coin probability")
plt.xlim([0,1])
plt.show()
```
