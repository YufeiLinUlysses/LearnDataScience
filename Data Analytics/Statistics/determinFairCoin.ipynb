# Fair Coin

```python
# Get numpy and all other important modules
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
%matplotlib inline
# Create a fair coin
coin = np.arange(2)
coin
# Fetch 1
np.random.choice(coin)
# Fetch 2, with duplicates
np.random.choice(coin, size=2, replace=True)
# Fetch 2, without duplicates
np.random.choice(coin, size=2, replace=False)
# check possibility
sum(np.random.choice(coin,size=500))/500
# double check
doubleCheckC = np.array([sum(np.random.choice(coin,size=500))/500 for _ in range(1000)])
_ = plt.hist(doubleCheckC, bins='auto')
plt.title("Histogram of Fair Coin probability")
plt.xlim([0,1])
plt.show()
```

```python
# Unfair coin
unfair = np.array([0] + [1]*10)
unfair
# Fetch 1
np.random.choice(unfair)
# Fetch 2, with duplicates
np.random.choice(unfair, size=2, replace=True)
# Fetch 2, without duplicates
np.random.choice(unfair, size=2, replace=False)
# check possibility
sum(np.random.choice(unfair,size=500))/500
# double check
doubleCheckU = np.array([sum(np.random.choice(unfair,size=500))/500 for _ in range(1000)])
_ = plt.hist(doubleCheckU, bins='auto')
plt.title("Histogram of Unfair Coin probability")
plt.xlim([0,1])
plt.show()
```

```python
# Combining both graphs together
tips = sns.load_dataset("tips")

# Put fair and unfair coins into one pandas dataframe
combined = pd.DataFrame({
    "Chance": np.concatenate((doubleCheckC,doubleCheckU)),
    "Type": ["Fair"]*1000+["Unfair"]*1000
})

# use facet grid to make a comparison between graphs
grid = sns.FacetGrid(combined, row="Type")
grid.map(plt.hist, "Chance")
```
